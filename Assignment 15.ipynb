```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('database.sqlite')
conn
```

```python
pd.read_sql_query('select * from salaries', conn)
```

```python
def sql_query(M):
    return pd.read_sql_query(M,conn)
```

#### **1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')**

```python
M = '''select JobTitle, AVG(TotalPayBenefits) Avg_TotalPayBenefits
    from salaries
    where JobTitle not in ('Not Provided', 'Not provided')
    group by JobTitle
    order by Avg_TotalPayBenefits desc'''
sql_query(M)
```

#### **2. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000**

```python
M = '''select jobtitle , overtimepay from salaries
    where overtimepay > 50000'''
sql_query(M)
```

#### **3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')**

```python
M = '''select distinct(employeename) from Salaries 
    where employeename != "Not provided" '''
sql_query(M)
```

#### **4.Show The Top 10 highest Paid Employees**

```python
M = '''select employeename , totalpay from salaries
    order by totalpay desc  limit 10'''

sql_query(M)
```

#### **5.Show All Employees Ordered By Their TotalPay Benefits In Descending Order**

```python
M = '''select EmployeeName, TotalPayBenefits
    from salaries
    order by TotalPayBenefits desc'''
sql_query(M)
```
